```python
def check_voter_eligibility():
    print("--- Voter Registration Eligibility Check ---")

    # Check Citizenship
    is_citizen = input("Are you a Kenyan citizen? (yes/no): ").lower().strip()

    if is_citizen != "yes":
        print("Sorry, only Kenyan citizens can register to vote.")
        return

    try:
        # Check Age
        age = int(input("Enter your age: "))

        if age >= 18:
            # Check for Identification Documents
            has_id = input("Do you have an original National ID or valid Passport? (yes/no): ").lower().strip()

            if has_id == "yes":
                print("You are eligible! Visit your nearest IEBC constituency office or Huduma Centre to register.")
            else:
                print("You must have a valid ID or Passport to register as a voter.")
        else:
            print(f"You are not of legal voting age yet. Please wait {18 - age} more year(s).")

    except ValueError:
        print("Invalid input! Please enter a number for your age.")

# Run the function
check_voter_eligibility()
```

```python
path = "/content/drive/MyDrive/titanic.csv"
```

```python
df=pd.read_csv(path)
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Age'] = df.groupby(['Pclass', 'Gender'])['Age'].transform(
    lambda x: x.fillna(x.median())
)

df['Cabin_known'] = df['Cabin'].notnull().astype(int)


df['Cabin'] = df['Cabin'].fillna("Unknown")


df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

print(df.isnull().sum())
```

```python
df.columns=df.columns.str.lower()
```

```python
df
```

```python
# survival rate

survival_rate=df["survived"].value_counts(normalize=True)*100
survival_rate.plot(kind="bar")
plt.title("survival percentage")
plt.ylabel("percentage")
plt.show()
```

```python
sns.boxplot(x="pclass",y="fare",data=df)
plt.title("fare by class")
plt.show
```

```python
df["family_size"]=df["sibsp"]+df["parch"]
sns.barplot(x="family_size",y="survived",data=df)
plt.title("Survival rate by family size")
plt.show()
```

```python
sns.histplot(df["age"],color="red")
plt.title("Age distribution")
plt.show()
```

```python
sns.histplot(data=df,x="gender",hue="survived",palette=["red","yellow"])
plt.title("gender by Survival")
plt.show
```

```python
survival=df["survived"].value_counts()
plt.pie(survival,labels=["Not Survived","Survived"],autopct="%1.1f%%",colors=["blue","green"])
plt.title("Survival %")
plt.show()
```

```python

min_fare_count = (df['fare'] == df['fare'].min()).sum()
max_fare_count = (df['fare'] == df['fare'].max()).sum()

fare_counts = [min_fare_count, max_fare_count]
labels = ["Minimum", "Maximum"]


plt.pie(fare_counts, labels=labels, autopct="%1.1f%%", colors=["red","blue"])
plt.title("Fare Distribution (Min vs Max)")
plt.show()
```

```python
sns.scatterplot(x="age",y="fare", data=df)
plt.title("Age vs Fare")
plt.show()
```

```python
sns.scatterplot(x="age",y="survived",data=df)
plt.title("Age vs Survived")
plt.show()
```
